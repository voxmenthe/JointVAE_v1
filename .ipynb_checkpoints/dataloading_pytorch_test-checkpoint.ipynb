```python
import numpy as np
import json
import cv2
import matplotlib.pyplot as plt

import glob
import numpy as np
from skimage.io import imread
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
```

```python
def filelist_to_data(files):
    '''
    __getitem__ creates a list of files to load for this batch.


    This method ingests that list and actually loads the required files,
    formats them as needed, and returns the read in files as consumable
    data.
    '''
    data = np.array([self.load_image(file) for file in files])

    return data

dim = (260,260) # (216, 261)
n_channels = 3
shape = (260,260,3) # (216, 261, 3)

def load_image(img_path):
    '''
    Load the image at the given path, process into a square image of the right size,
    scale the color values to float 0.0 - 1.0 and return as numpy array
    '''

    root_path = "/home/jovyan/vmldata/raw_source_data/v20180810_all_wearables"

    try:
        # pylint: disable=no-member
        img = cv2.imread(root_path + img_path)
    except:
        print("unable to read: ",root_path+img_path)
        img = None

    # seems like the image failed to properly load so work with a fake noise image
    # the distance will be huge but it won't be zeros and make the model think that
    # it's really good
    if img is None or img.shape[2] != 3 or len(img.shape) != 3:  
        print("img failed to load: img_path:", root_path + img_path)
        img = np.random.randint(0, 256, size=(dim[0], dim[1], n_channels), dtype=np.uint8)

    img = img[..., ::-1]  # reverse color channels to be correct

    # figure out if the image needs to be scaled and if so which direction/magnitude
    scale = shape[0] / max(img.shape[:2])
    if scale != 1:
        # pylint: disable=no-member
        img = cv2.resize(img, dsize=None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)

    # matrix to hold full (square) image:
    final_img = np.zeros(shape, dtype=np.uint8)

    # whiten image because most Macy's backgrounds are nearly white studio
    final_img += 255

    # for macys image data typically the image is a vertically tall rectangle so float the image
    # in the middle of padding, randomly
    x_offset = np.random.randint(0, final_img.shape[0] - img.shape[0] + 1)
    y_offset = np.random.randint(0, final_img.shape[1] - img.shape[1] + 1)
    final_img[x_offset:x_offset + img.shape[0], y_offset:y_offset + img.shape[1], :] = img

    # scale image to float, might be best to rescale this more so
    # the data is centered at 0 (ie -1. to 1.) but not sure with pytorch
    # implementation
    final_img = final_img.astype(np.float32)

    final_img = final_img / 255.

    final_img[..., 0] -= (103.939 / 255.0)    # red
    final_img[..., 1] -= (116.779 / 255.0)    # green
    final_img[..., 2] -= (123.68 / 255.0)     # blue

    return final_img
```
